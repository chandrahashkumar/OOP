```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [20,10,20,45,30]
plt.plot(x,y)
plt.show()
```

```python
x =['Monday','Tuesday','Wednesday','Thursday','Friday']
y =[12,15,20,8,12]
plt.plot(x,y)
plt.title("Bakery Sales This week! ")
plt.xlabel("Day of the week")
plt.ylabel("Sales per day")

plt.show()
```

```python
months =[1,2,3,4]
sales = [3000,3500,8000,10000]
plt.plot(months,sales,color='blue',linestyle='--',linewidth =2, marker = 's',label='2025 sales data')

plt.xlabel("Months")
plt.ylabel("Sales per month")
plt.title("Monthly Sales Data Report")
plt.legend(loc='upper left',fontsize=12)
plt.grid(color='gray',linestyle=':',linewidth =1)
plt.xlim(1,4)
plt.ylim(0,8000)
plt.xticks([1,2,3,4],['M1','M2','M3','M4'])
plt.show()
```

```python
years =[1,2,3,4]
cgp = [8.91,9.2,9.5,9.7]
plt.plot(years,cgp,color ='yellow',linestyle="--",linewidth =2,label ="CGP",marker ='o')
plt.xlabel("Year")
plt.ylabel("CGP")
plt.title("B.Tech AI&ML")
plt.grid(color ='green',linestyle ='-.',linewidth =1)
plt.legend(loc ='upper left',fontsize =13)
plt.xticks([1,2,3,4],['1 year','2 year','3 year','4 year'])
plt.show()


```

```python
product =['a','b','c','d']
sales = [100000,150000,100000,400000]
plt.bar(product,sales,color="red",label='Sales 2025')
plt.xlabel("Product")
plt.ylabel("Sales")
plt.title("Product sales comparison")
plt.legend()
plt.show()
```

```python
regions =['North','South','East','West']
revenue =[40000,30000,45000,100000]
plt.pie(revenue,labels =regions,autopct='%1.1f%%',colors=['green','red','skyblue','gold'])
plt.title("Revenue Contribution By Regions")
plt.show()
```

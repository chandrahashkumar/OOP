```python
months =[1,2,3,4]
sales = [3000,3500,8000,10000]
plt.plot(months,sales,color='blue',linestyle='--',linewidth =2, marker = 's',label='2025 sales data')

plt.xlabel("Months")
plt.ylabel("Sales per month")
plt.title("Monthly Sales Data Report")
plt.legend(loc='upper left',fontsize=12)
plt.grid(color='gray',linestyle=':',linewidth =1)
plt.xlim(1,4)
plt.ylim(0,8000)
plt.xticks([1,2,3,4],['M1','M2','M3','M4'])
plt.show()
```

```python
years =[1,2,3,4]
cgp = [8.91,9.2,9.5,9.7]
plt.plot(years,cgp,color ='yellow',linestyle="--",linewidth =2,label ="CGP",marker ='o')
plt.xlabel("Year")
plt.ylabel("CGP")
plt.title("B.Tech AI&ML")
plt.grid(color ='green',linestyle ='-.',linewidth =1)
plt.legend(loc ='upper left',fontsize =13)
plt.xticks([1,2,3,4],['1 year','2 year','3 year','4 year'])
plt.show()


```

```python
hours_studies =[1,2,3,4,5,6,7,8]
exam_scores =[50,60,68,80,98,99,98,99]
plt.scatter(hours_studies,exam_scores,color='red',marker ='D',label ='Student Data')
plt.xlabel("Hours studies")
plt.ylabel("Exam Score")
plt.title("Relationship Between Study and Exam Score")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.scatter([1,2,3,4,5,6,7],[55,63,68,70,80,90,98],color='blue',marker='s',label="Section A")
plt.scatter([1,2,3,4,5,6,7],[45,50,60,68,75,85,90],color='yellow',marker='o',label="Section B")
plt.xlabel("Hours Studies")
plt.ylabel("Exam Score")
plt.title("Relationship Between Study Hours and Exam Score")
plt.legend()
plt.grid()
plt.savefig("scatter_simple.png",dpi=300,bbox_inches="tight")
plt.show(True)
```

```python
x =[1,2,3,4]
y = [10,20,30,35]
plt.subplot(1,2,1) # 1st_row 2nd_column 1st_subplot
plt.plot(x,y,color="green")
plt.title("Line Chart")

plt.subplot(1,2,2)
plt.bar(x,y,color="gold")
plt.title("Bar Chart")

plt.tight_layout()
plt.show()
```

```python
#fig, ax = plt.subplots(n_row,n_column,figsize=(width, height))

fig, ax = plt.subplots(1,2, figsize =(10,5))
x =[1,2,3,4]
y = [10,20,30,35]

ax[0].plot(x,y,color="blue")
ax[0].set_title("Line Plot")

ax[1].bar(x,y,color="green")
ax[1].set_title("Line Chart")

fig.suptitle("Compersion of Line and Bar Chart")
plt.tight_layout()
plt.savefig('compersion_chart.pdf',dpi=300,bbox_inches ='tight')
plt.show()
```

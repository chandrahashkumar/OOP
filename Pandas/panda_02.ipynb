Merging and join
---pd.merge(df1,df2,on="Column_Name",how = "Type of join")

```python
df_customers=pd.DataFrame({"Customer_ID":[1,2,3,4,5],
                           "Name":['Rohit','Shyam','AAKASH','Jone','Janni']})
df_order =pd.DataFrame({"Customer_ID":[1,2,4,5,6],
                        "Order_Amount":[2500,600,4500,4510,56494]})
```

```python
df_merge =pd.merge(df_customers,df_order,on='Customer_ID',how="inner")
df_merge
```

```python
df_outer =pd.merge(df_customers,df_order,on='Customer_ID',how="outer")
df_outer
```

```python
df_left =pd.merge(df_customers,df_order,on='Customer_ID',how="left")
df_left
```

```python
df_right =pd.merge(df_customers,df_order,on='Customer_ID',how="right")
df_right
```
